```python
import cv2
import os

# Define paths
data_dir = 'dataset'
categories = [chr(i) for i in range(65, 91)]  # 'A' to 'Z'
categories.append('Space')  # Add space to categories

# Function to capture and save images
def capture_images():
    cap = cv2.VideoCapture(0)
    
    # Capture images for A-Z
    for category in categories[:-1]:  # Exclude 'Space' here
        path = os.path.join(data_dir, category)
        if not os.path.exists(path):
            os.makedirs(path)

        print(f"Press 'q' to start capturing images for sign '{category}'...")
        # Wait for the user to press 'q' to start capturing
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Failed to capture image")
                break

            # Define Region of Interest (ROI) for the hand gesture
            roi = frame[100:300, 100:300]
            gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            
            # Show the frame and ROI on the screen
            cv2.imshow(f'Waiting to capture images for {category}', gray)
            
            # Start capturing when 'q' is pressed
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        
        # Collect 100 images for each letter
        count = 0
        while count < 100:
            ret, frame = cap.read()
            if not ret:
                print("Failed to capture image")
                break
            
            roi = frame[100:300, 100:300]
            gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
            
            # Show the capturing frame and ROI on the screen
            cv2.imshow(f'Capturing images for {category}', gray)
            
            # Save the image when 'c' is pressed or automatically save each frame
            img_name = f"{category}_{count}.png"
            img_path = os.path.join(path, img_name)
            cv2.imwrite(img_path, gray)
            count += 1
            print(f"Captured {count}/100 for {category}")
            
            if cv2.waitKey(1) & 0xFF == ord('q'):
                print(f"Finished capturing images for '{category}'")
                break
    
    # Capture images for the space gesture
    
    13
    space_path = os.path.join(data_dir, 'Space')
    if not os.path.exists(space_path):
        os.makedirs(space_path)

    print("Press 'q' to start capturing images for sign 'Space'...")
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break

        # Define Region of Interest (ROI) for the hand gesture
        roi = frame[100:300, 100:300]
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        
        # Show the frame and ROI on the screen
        cv2.imshow('Waiting to capture images for Space', gray)
        
        # Start capturing when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Collect 100 images for space
    count = 0
    while count < 100:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        
        roi = frame[100:300, 100:300]
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        
        # Show the capturing frame and ROI on the screen
        cv2.imshow('Capturing images for Space', gray)
        
        # Save the image
        img_name = f"Space_{count}.png"
        img_path = os.path.join(space_path, img_name)
        cv2.imwrite(img_path, gray)
        count += 1
        print(f"Captured {count}/100 for Space")
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Finished capturing images for 'Space'")
            break
        
    cap.release()
    cv2.destroyAllWindows()

# Start the data collection process
if __name__ == "__main__":
    capture_images()
```
