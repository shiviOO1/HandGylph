```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model

# Load the trained model
model = load_model('sign_language_model.h5')

# Define class names
categories = [chr(i) for i in range(65, 91)] + ['Space']  # 'A' to 'Z' and 'Space'

# Function to predict sign language
def predict_sign_language():
    cap = cv2.VideoCapture(0)
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to capture image")
            break
        
        roi = frame[100:300, 100:300]
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
        gray = cv2.resize(gray, (64, 64))
        gray = gray.reshape(-1, 64, 64, 1).astype('float32') / 255.0  # Normalize

        prediction = model.predict(gray)
        predicted_class = np.argmax(prediction, axis=-1)[0]
        predicted_letter = categories[predicted_class]

        # Display the prediction
        cv2.putText(frame, f'Predicted: {predicted_letter}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
        cv2.rectangle(frame, (100, 100), (300, 300), (0, 255, 0), 2)
        cv2.imshow('Sign Language Prediction', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

# Start prediction process
if __name__ == "__main__":
    predict_sign_language()
```
